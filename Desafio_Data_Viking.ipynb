*Passo 4 - Análise das causas do turnover (como as colunas da base impactam no desligamento) e criar os gráficos para análise*

```python
import plotly.express as px

df = df.drop(columns="Data_Desligamento")

for coluna in df.columns:

  grafico = px.histogram(df, x=coluna, color="Desligamento")

  grafico.update_traces(textposition='inside',texttemplate='%')
  grafico.update_yaxes(showticklabels=False)
  grafico.update_layout(title={
    'y': 0.9,
    'x': 0.5
  })

  grafico.show()
```
